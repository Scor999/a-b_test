## Решение ##
На основе задания из второго урока было принято решение по итогам ICE, составить шаблоны эксперимента для двух гипотез набравших максимальное количество очков:
+ Организация программы лояльности персональными скидками для постоянных клиентов увеличит количество повторных покупок - 124 балла
+  Внедрение системы рекомендаций товаров на основе истории покупок и предпочтений клиентов повысит конверсию посетителей в покупателей - 84 балла

### 1. Шаблон А/B-теста добавить пункт «Организация программы лояльности персональными скидками для постоянных клиентов» ###

#### Гипотеза ####
Если организовать программы лояльности с персональными скидками для постоянных клиентов, то % количества повторных покупок увеличиться на 3-5%.

#### Что делаем.
Контрольная версия: оставляем текущий вид страницы.                 
Тестовая версия: Добавляем всплывающее окно с персональными предложениями по интересам клиента                      
На каких пользователях тестируем.                       
На новых пользователях                     
Ожидаемый эффект.                   
Увеличение среднего чека за счет дополнительных покупок
Метрики:
+ Процент конверсии в заказ (%CR) — primary. 
+ Количество новых пользователей — secondary.                          
                      


План действий в зависимости от результатов эксперимента.                        
В случае успешного проведения эксперимента, и повышения метрики по среднему чеку, добавляем функционал для всех пользователей.                          
Если показатели тестирования покажет не высокую динамику, отмена эксперимента.

#### Что делаем.
Контрольная версия: оставляем текущий вид страницы.                 
Тестовая версия: Внедрение системы рекомендаций товаров на основе истории покупок и предпочтений клиентов                      
На каких пользователях тестируем.                       
На старых пользователях                    
Ожидаемый эффект.                   
Увеличение среднего чека за счет дополнительных покупок
Метрики:
+ Процент конверсии в заказ (%CR)— primary. 
+ Количество новых пользователей — secondary.                          


План действий в зависимости от результатов эксперимента.
